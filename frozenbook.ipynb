```python
from dataclasses import dataclass

@dataclass(frozen=True)
class frozenbook:
    tile : str

fb1 = frozenbook("fb1")
print(frozenbook.__dataclass_params__)
frozenbook.__dataclass_params__.frozen = False
print(frozenbook.__dataclass_params__)

fb2 = frozenbook("dhadkh")
fb2.title = "hello"
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data,bins=30)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequencey")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000)
y = np.random.randn(1000)

plt.hist2d(x,y,bins=20)
plt.title("Histogram")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="frequence")
```

```python
import matplotlib.pyplot as plt
x =["A","B","C","D"]
y=[10,20,15,30]
plt.pie(y,labels=x,autopct="%1.1f%%")
plt.title("Pie plot")
plt.xlabel("Xaxis")
plt.ylabel("Yaxis")
plt.show()
```
